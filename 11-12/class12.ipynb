## MERGE

```python
leftDataFrame = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'KL'],

                     'A': ['A0', 'A1', 'A2', 'A3'],

                     'B': ['B0', 'B1', 'B2', 'B3']})

   

rightDataFrame = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'KR'],

                          'C': ['C0', 'C1', 'C2', 'C3'],

                          'D': ['D0', 'D1', 'D2', 'D3']})    
```

```python
leftDataFrame
```

```python
rightDataFrame
```

```python
pd.merge(leftDataFrame, rightDataFrame, how='inner', on='key')
```

```python
pd.merge(leftDataFrame, rightDataFrame, how='outer', on='key')
```

```python
pd.merge(leftDataFrame, rightDataFrame, how='left', on='key')
```

```python
pd.merge(leftDataFrame, rightDataFrame, how='right', on='key')
```

## JOIN

```python
leftDataFrame.merge(rightDataFrame, how='outer')
```
