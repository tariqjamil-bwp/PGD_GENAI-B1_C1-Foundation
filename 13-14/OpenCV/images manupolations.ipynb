Let's create a simple function to make displaying our images simpler and easier

```python
def imshow(title = "", image = None):
    if len(image.shape)==3:
        image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    plt.imshow(image)
    plt.title(title)
    plt.show()
```

```python
image2 = cv2.imread('building.jpeg')
```

```python
# Slicing and indexing
img = image2.copy()
red_channel = img[:, :, 0]  # Extract red channel
grayscale_img = 0.299 * red_channel + 0.587 * img[:, :, 1] + 0.114 * img[:, :, 2]  # Convert to grayscale
```

```python
plt.imshow(grayscale_img)
```

```python

# Image transformations
resized_img = np.resize(img, (256, 256))  # Resize image
rotated_img = np.rot90(img, 1)  # Rotate image 90 degrees clockwise

# Applying filters
blur_kernel = np.ones((3, 3)) / 9
blurred_img = np.convolve(img, blur_kernel, mode='same')

# Saving the modified image
Image.fromarray(resized_img).save("resized_image.jpg")

# Numerical operations
mean_intensity = np.mean(img)
standard_deviation = np.std(img)

# Displaying images with matplotlib
import matplotlib.pyplot as plt

plt.subplot(221)
plt.imshow(img)
plt.title("Original image")

plt.subplot(222)
plt.imshow(grayscale_img, cmap="gray")
plt.title("Grayscale image")

plt.subplot(223)
plt.imshow(rotated_img)
plt.title("Rotated image")

plt.subplot(224)
plt.imshow(blurred_img)
plt.title("Blurred image")

plt.show()
```
